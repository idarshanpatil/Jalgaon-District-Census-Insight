```python
# Remove irrelevant columns (e.g., administrative codes)
columns_to_drop = ['State Code', 'District Code', 'Sub-Distt', 'Town/ Village Code', 'Ward Code', 'Enumeration Block Code']
df = df.drop(columns=columns_to_drop, errors='ignore')

# Rename columns for clarity
df = df.rename(columns={
    'Total Population - Persons': 'Total Population',
    'Total Population - Males': 'Male Population',
    'Total Population - Females': 'Female Population',
    'Population in the Age Group 0-6 - Persons': 'Population 0-6',
    'Population in the Age Group 0-6 - Males': 'Male Population 0-6',
    'Population in the Age Group 0-6 Females': 'Female Population 0-6',
    'Scheduled Castes Population - Persons': 'Scheduled Castes Population',
    'Scheduled Castes Population - Males': 'Male Scheduled Castes Population',
    'Scheduled Castes Population - Females': 'Female Scheduled Castes Population',
    'Scheduled Tribes Population - Persons': 'Scheduled Tribes Population',
    'Scheduled Tribes Population - Males': 'Male Scheduled Tribes Population',
    'Scheduled Tribes Population - Females': 'Female Scheduled Tribes Population',
    'Literates Population - Persons': 'Literate Population',
    'Literates Population - Males': 'Male Literate Population',
    'Literates Population - Females': 'Female Literate Population',
    'Illiterate - Persons': 'Illiterate Population',
    'Illiterate - Males': 'Male Illiterate Population',
    'Illiterate - Females': 'Female Illiterate Population',
    'Total Worker Population - Persons': 'Total Worker Population',
    'Total Worker Population - Males': 'Male Worker Population',
    'Total Worker Population - Females': 'Female Worker Population',
    'Main Worker Population - Persons': 'Main Worker Population',
    'Main Worker Population - Males': 'Male Main Worker Population',
    'Main Worker Population - Females': 'Female Main Worker Population',
    'Main Cultivator Population - Persons': 'Main Cultivator Population',
    'Main Cultivator Population - Males': 'Male Main Cultivator Population',
    'Main Cultivator Population - Females': 'Female Main Cultivator Population',
    'Main Agricultural Labourers Population - Persons': 'Main Agricultural Laborer Population',
    'Main Agricultural Labourers Population - Males': 'Male Main Agricultural Laborer Population',
    'Main Agricultural Labourers Population - Females': 'Female Main Agricultural Laborer Population',
    'Main Household Industries Population - Persons': 'Main Household Industries Population',
    'Main Household Industries Population - Males': 'Male Main Household Industries Population',
    'Main Household Industries Population - Females': 'Female Main Household Industries Population',
    'Main Other Workers Population - Persons': 'Main Other Workers Population',
    'Main Other Workers Population - Males': 'Male Main Other Workers Population',
    'Main Other Workers Population - Females': 'Female Main Other Workers Population',
    'Marginal Worker Population - Persons': 'Marginal Worker Population',
    'Marginal Worker Population - Males': 'Male Marginal Worker Population',
    'Marginal Worker Population - Females': 'Female Marginal Worker Population',
    'Marginal Cultivator Population - Persons': 'Marginal Cultivator Population',
    'Marginal Cultivator Population - Males': 'Male Marginal Cultivator Population',
    'Marginal Cultivator Population - Females': 'Female Marginal Cultivator Population',
    'Marginal Agriculture Labourers Population - Persons': 'Marginal Agricultural Laborer Population',
    'Marginal Agriculture Labourers Population - Males': 'Male Marginal Agricultural Laborer Population',
    'Marginal Agriculture Labourers Population - Females': 'Female Marginal Agricultural Laborer Population',
    'Marginal Household Industries Population - Persons': 'Marginal Household Industries Population',
    'Marginal Household Industries Population - Males': 'Male Marginal Household Industries Population',
    'Marginal Household Industries Population - Females': 'Female Marginal Household Industries Population',
    'Marginal Other Workers Population - Persons': 'Marginal Other Workers Population',
    'Marginal Other Workers Population - Males': 'Male Marginal Other Workers Population',
    'Marginal Other Workers Population - Females': 'Female Marginal Other Workers Population',
    'Non Worker Population - Persons': 'Non Worker Population',
    'Non Worker Population - Males': 'Male Non Worker Population',
    'Non Worker Population - Females': 'Female Non Worker Population',
})

# Check for and remove duplicate rows
df = df.drop_duplicates()

display(df.head())
```

```python
# Ensure numerical columns are of the correct data type
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        df[column] = pd.to_numeric(df[column], errors='coerce')

display(df.info())
```

```python
# Create age groups and calculate proportions
df['Population 0-14'] = df['Population 0-6']  # Assuming 'Population 0-6' is a reasonable proxy for 0-14
df['Population 15-64'] = df['Total Population'] - df['Population 0-14'] - df['Non Worker Population']
df['Population 65+'] = df['Non Worker Population']

df['Proportion 0-14'] = (df['Population 0-14'] / df['Total Population']) * 100
df['Proportion 15-64'] = (df['Population 15-64'] / df['Total Population']) * 100
df['Proportion 65+'] = (df['Population 65+'] / df['Total Population']) * 100

display(df[['Total Population', 'Population 0-14', 'Population 15-64', 'Population 65+', 'Proportion 0-14', 'Proportion 15-64', 'Proportion 65+']].head())
```

```python
# Categorize employment types
df['Agriculture Population'] = df['Main Cultivator Population'] + df['Main Agricultural Laborer Population'] + df['Marginal Cultivator Population'] + df['Marginal Agricultural Laborer Population']
df['Industry Population'] = df['Main Household Industries Population'] + df['Marginal Household Industries Population']
df['Services Population'] = df['Main Other Workers Population'] + df['Marginal Other Workers Population']

df['Proportion Agriculture'] = (df['Agriculture Population'] / df['Total Worker Population']) * 100
df['Proportion Industry'] = (df['Industry Population'] / df['Total Worker Population']) * 100
df['Proportion Services'] = (df['Services Population'] / df['Total Worker Population']) * 100

display(df[['Total Worker Population', 'Agriculture Population', 'Industry Population', 'Services Population', 'Proportion Agriculture', 'Proportion Industry', 'Proportion Services']].head())
```

```python
# Identify urban/rural areas
df['Area Type'] = 'Rural'
df.loc[df['Total/ Rural/ Urban'] == 'Urban', 'Area Type'] = 'Urban'

display(df[['Total/ Rural/ Urban', 'Area Type']].head())
```

```python
total_population = df.loc[df['Level'] == 'DISTRICT', 'Total Population'].iloc[0]
print(f"Total population of Jalgaon district: {total_population}")

# Calculate urban vs. rural population proportions
urban_population = df.loc[(df['Level'] == 'DISTRICT') & (df['Total/ Rural/ Urban'] == 'Urban'), 'Total Population'].sum()
rural_population = df.loc[(df['Level'] == 'DISTRICT') & (df['Total/ Rural/ Urban'] == 'Rural'), 'Total Population'].sum()
urban_proportion = (urban_population / total_population) * 100
rural_proportion = (rural_population / total_population) * 100

print(f"Urban population proportion: {urban_proportion:.2f}%")
print(f"Rural population proportion: {rural_proportion:.2f}%")
```

```python
# Calculate the sex ratio
district_df = df[df['Level'] == 'DISTRICT']
male_population = district_df['Male Population'].iloc[0]
female_population = district_df['Female Population'].iloc[0]

sex_ratio = (female_population / male_population) * 1000

print(f"Sex ratio (females per 1,000 males): {sex_ratio:.2f}")
```

```python
# Calculate employment rates in different sectors
district_df = df[df['Level'] == 'DISTRICT']
total_worker_population = district_df['Total Worker Population'].iloc[0]

agriculture_employment_rate = (district_df['Agriculture Population'].iloc[0] / total_worker_population) * 100
industry_employment_rate = (district_df['Industry Population'].iloc[0] / total_worker_population) * 100
services_employment_rate = (district_df['Services Population'].iloc[0] / total_worker_population) * 100

print(f"Agriculture employment rate: {agriculture_employment_rate:.2f}%")
print(f"Industry employment rate: {industry_employment_rate:.2f}%")
print(f"Services employment rate: {services_employment_rate:.2f}%")

```

```python
# Calculate average household size
district_df = df[df['Level'] == 'DISTRICT']
total_population = district_df['Total Population'].iloc[0]
no_of_households = district_df['No of Households'].iloc[0]

average_household_size = total_population / no_of_households

print(f"Average household size: {average_household_size:.2f}")
```

```python
import matplotlib.pyplot as plt

# 1. Population Distribution: Urban vs. Rural
district_df = df[df['Level'] == 'DISTRICT']
urban_population = df.loc[(df['Level'] == 'DISTRICT') & (df['Total/ Rural/ Urban'] == 'Urban'), 'Total Population'].sum()
rural_population = df.loc[(df['Level'] == 'DISTRICT') & (df['Total/ Rural/ Urban'] == 'Rural'), 'Total Population'].sum()

labels = ['Urban', 'Rural']
sizes = [urban_population, rural_population]
colors = ['lightcoral', 'lightskyblue']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Population Distribution: Urban vs. Rural')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# 2. Age Structure: Proportion of different age groups
district_df = df[df['Level'] == 'DISTRICT']
population_0_14 = district_df['Population 0-14'].iloc[0]
population_15_64 = district_df['Population 15-64'].iloc[0]
population_65_plus = district_df['Population 65+'].iloc[0]

labels = ['0-14', '15-64', '65+']
sizes = [population_0_14, population_15_64, population_65_plus]
colors = ['gold', 'lightgreen', 'lightcoral']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Age Structure of the Population')
plt.axis('equal')
plt.show()
```

```python
# 3. Sex Ratio: Display the sex ratio as a bar chart
district_df = df[df['Level'] == 'DISTRICT']
male_population = district_df['Male Population'].iloc[0]
female_population = district_df['Female Population'].iloc[0]
sex_ratio = (female_population / male_population) * 1000

plt.figure(figsize=(6, 4))
plt.bar(['Sex Ratio'], [sex_ratio], color='skyblue')
plt.title('Sex Ratio (Females per 1000 Males)')
plt.ylabel('Ratio')
plt.show()
```

```python
# 4. Employment Sectors: Proportion of employment in different sectors
district_df = df[df['Level'] == 'DISTRICT']
total_worker_population = district_df['Total Worker Population'].iloc[0]

agriculture_population = district_df['Agriculture Population'].iloc[0]
industry_population = district_df['Industry Population'].iloc[0]
services_population = district_df['Services Population'].iloc[0]

labels = ['Agriculture', 'Industry', 'Services']
sizes = [agriculture_population, industry_population, services_population]
colors = ['forestgreen', 'slategrey', 'steelblue']

plt.figure(figsize=(8, 5))
plt.bar(labels, [(size/total_worker_population)*100 for size in sizes], color=colors)
plt.title('Proportion of Employment in Different Sectors')
plt.ylabel('Percentage')
plt.show()
```

```python
# 5. Literacy Rates: Compare overall, male, and female literacy rates
district_df = df[df['Level'] == 'DISTRICT']
overall_literacy_rate = district_df['Overall Literacy Rate'].iloc[0]
male_literacy_rate = district_df['Male Literacy Rate'].iloc[0]
female_literacy_rate = district_df['Female Literacy Rate'].iloc[0]

labels = ['Overall', 'Male', 'Female']
sizes = [overall_literacy_rate, male_literacy_rate, female_literacy_rate]
colors = ['skyblue', 'royalblue', 'lightblue']

plt.figure(figsize=(8, 5))
plt.bar(labels, sizes, color=colors)
plt.title('Literacy Rates')
plt.ylabel('Percentage')
plt.show()
```

```python
# 6. Average Household Size: Display the average household size
district_df = df[df['Level'] == 'DISTRICT']
total_population = district_df['Total Population'].iloc[0]
no_of_households = district_df['No of Households'].iloc[0]
average_household_size = total_population / no_of_households

plt.figure(figsize=(6, 4))
plt.bar(['Average Household Size'], [average_household_size], color='lightgreen')
plt.title('Average Household Size')
plt.ylabel('Size')
plt.show()
```
